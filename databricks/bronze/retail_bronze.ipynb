```python
root = '/Volumes/s3_data_platform/s3_data/raw/retail/'
tables = [
    'categories',
    'customers',
    'departments',
    'order_items',
    'orders',
    'products'
]
```

```python
def get_raw_table(table):
    table = f"{root}{table}/"
    return spark.read.csv(table, header=False, inferSchema=True)
```

```python
def save_bronze_table(df,table):
    df.write.mode("overwrite").option("overwriteSchema", "true").saveAsTable(f'bronze.retail.{table}', format='delta')
```

```python
for table in tables:
    df = get_raw_table(table)
    save_bronze_table(df, table)
```
